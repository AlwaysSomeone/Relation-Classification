```python
# Dependency Tree
from nltk.parse.stanford import StanfordDependencyParser
dep_parser=StanfordDependencyParser(model_path="/home/shanu/nltk/jars/englishPCFG.ser.gz")
```

```python
sentences= "This strawberry sauce has about a million uses , is freezer-friendly , and is so much better than that jar of Smuckers strawberry sauce that you 've had sitting in your fridge since that time you made banana splits 1.5 years ago ."
```

```python
from nltk.tokenize.stanford import StanfordTokenizer
path_to_jar = "/home/shanu/nltk/jars/stanford-postagger.jar"
tokenizer=StanfordTokenizer(path_to_jar)
```

```python
tokenizer.tokenize(sentences)
```

```python
 parse_tree = dep_parser.raw_parse(sentences)
for trees in parse_tree:
    tree = trees
    t = trees.tree()
```

```python
print(tree)
```

```python
for i in tree.nodes:
    print(tree.nodes[i]['word'], i)
```

```python
for p in t:
    print(p.tree())
```
